```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("london_weather.csv")
data.head()
```

```python
data
```

```python
data.iloc[100:200]
```

```python
max_temp = data["max_temp"]
mean_temp = data["mean_temp"]
min_temp = data["min_temp"]



average_temp = (max_temp + mean_temp + min_temp) / 3
# print(average_temp)

```

```python
data["average_temp"] = average_temp
data.head()
```

```python

max_temp = np.mean(data["max_temp"])
mean_temp = np.mean(data["mean_temp"])
min_temp = np.mean(data["min_temp"])
max_temp, mean_temp, min_temp
```

```python
std_temp = np.std(data["average_temp"])
std_temp
```
